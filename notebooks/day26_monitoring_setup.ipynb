```python
print("""
Day 26 - Monitoring Setup

Objective:
- Configure Prometheus
- Configure Grafana Dashboard
- Prepare Monitoring Infrastructure
- Enable KPI Tracking
""")
```

```python
import os

os.makedirs("../monitoring", exist_ok=True)

print("Monitoring folder created successfully")
```

```python
prometheus_config = """
global:
  scrape_interval: 15s

scrape_configs:
  - job_name: 'retailpulse'

    static_configs:
      - targets: ['localhost:8501']
"""

with open("../monitoring/prometheus.yml", "w") as f:
    f.write(prometheus_config)

print("prometheus.yml created")
```

```python
with open("../monitoring/prometheus.yml", "r") as f:
    print(f.read())
```

```python
grafana_dashboard = """
{
  "title": "RetailPulse Dashboard",

  "panels": [
    {
      "title": "Forecast Demand",
      "type": "graph"
    },
    {
      "title": "Customer Churn",
      "type": "graph"
    },
    {
      "title": "Inventory Status",
      "type": "graph"
    }
  ]
}
"""

with open("../monitoring/grafana_dashboard.json", "w") as f:
    f.write(grafana_dashboard)

print("grafana_dashboard.json created")
```

```python
with open("../monitoring/grafana_dashboard.json", "r") as f:
    print(f.read())
```

```python
import os

prometheus_exists = os.path.exists("../monitoring/prometheus.yml")
grafana_exists = os.path.exists("../monitoring/grafana_dashboard.json")

print("Prometheus Config Exists:", prometheus_exists)
print("Grafana Dashboard Exists:", grafana_exists)
```
