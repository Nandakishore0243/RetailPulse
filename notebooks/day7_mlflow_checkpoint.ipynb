```python
import pandas as pd
import numpy as np

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error
)

import mlflow
```

```python
prophet_df = pd.read_csv(
    "../data/processed/sales_forecast.csv"
)

prophet_df.head()
```

```python
lstm_df = pd.read_csv(
    "../data/processed/lstm_forecast.csv"
)

lstm_df.head()
```

```python
actual = lstm_df["Actual"]
predicted = lstm_df["Predicted"]

lstm_mae = mean_absolute_error(
    actual,
    predicted
)

lstm_rmse = np.sqrt(
    mean_squared_error(
        actual,
        predicted
    )
)

print("MAE :", lstm_mae)
print("RMSE:", lstm_rmse)
```

```python
with mlflow.start_run(run_name="LSTM_Forecast"):

    mlflow.log_param("Model", "LSTM")

    mlflow.log_metric("MAE", lstm_mae)
    mlflow.log_metric("RMSE", lstm_rmse)

print("LSTM metrics logged successfully!")
```
