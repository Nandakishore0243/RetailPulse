```python
import pandas as pd

forecast_df = pd.read_csv(
    "../data/processed/hybrid_forecast.csv"
)

customer_df = pd.read_csv(
    "../data/processed/customer_segments.csv"
)

inventory_df = pd.read_csv(
    "../data/processed/inventory_optimization.csv"
)

print("Data Loaded Successfully")
```

```python
reorder_count = (
    inventory_df["Inventory_Status"]
    == "Reorder Required"
).sum()

print("Products needing reorder:", reorder_count)
```

```python
predicted_churn = int(
    len(customer_df) * 0.10
)

print(
    "Predicted churn customers:",
    predicted_churn
)
```

```python
avg_forecast = forecast_df["Hybrid"].mean()

print(
    "Average Forecast Demand:",
    round(avg_forecast, 2)
)
```
