```python
import pandas as pd
import numpy as np
```

```python
forecast_df = pd.read_csv("../data/processed/hybrid_forecast.csv")

forecast_df.head()
```

```python
mae = np.mean(
    np.abs(
        forecast_df["Actual"] -
        forecast_df["Hybrid"]
    )
)

print("MAE:", round(mae, 2))
```

```python
rmse = np.sqrt(
    np.mean(
        (
            forecast_df["Actual"] -
            forecast_df["Hybrid"]
        ) ** 2
    )
)

print("RMSE:", round(rmse, 2))
```

```python
accuracy = 100 - (
    mae /
    forecast_df["Actual"].mean()
) * 100

print(
    "Forecast Accuracy:",
    round(accuracy, 2),
    "%"
)
```

```python
import time

start = time.time()

for i in range(10000):
    _ = forecast_df["Hybrid"].mean()

end = time.time()

print(
    "Load Test Time:",
    round(end - start, 4),
    "seconds"
)
```

```python
large_df = pd.concat(
    [forecast_df] * 100,
    ignore_index=True
)

print(
    "Stress Test Dataset Size:",
    large_df.shape
)
```

```python
validation = {
    "MAE": round(mae, 2),
    "RMSE": round(rmse, 2),
    "Accuracy": round(accuracy, 2)
}

print(validation)
```
