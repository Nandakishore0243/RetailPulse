```python
import pandas as pd

inventory_df = pd.read_csv(
    "../data/processed/inventory_optimization.csv"
)

inventory_df.head()
```

```python
inventory_df["Inventory_Status"].value_counts()
```

```python
import matplotlib.pyplot as plt

inventory_df["Inventory_Status"].value_counts().plot(
    kind="bar"
)

plt.title("Inventory Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

```python
reorder_count = (
    inventory_df["Inventory_Status"]
    == "Reorder Required"
).sum()

sufficient_count = (
    inventory_df["Inventory_Status"]
    == "Stock Sufficient"
).sum()

print("Reorder Required:", reorder_count)
print("Stock Sufficient:", sufficient_count)
```
