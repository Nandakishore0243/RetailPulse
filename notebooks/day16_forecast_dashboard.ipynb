```python
import pandas as pd

forecast_df = pd.read_csv("../data/processed/hybrid_forecast.csv")

forecast_df.head()
```

```python
print("Shape:", forecast_df.shape)
print("\nColumns:")
print(forecast_df.columns)
```

```python
forecast_df.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

plt.plot(forecast_df["Actual"], label="Actual")
plt.plot(forecast_df["Prophet"], label="Prophet")
plt.plot(forecast_df["LSTM"], label="LSTM")
plt.plot(forecast_df["Hybrid"], label="Hybrid")

plt.title("Forecast Comparison")
plt.legend()
plt.show()
```

```python
models = ["Hybrid","LSTM","Prophet"]
mae_values = [
    16030.93,
    14078.20,
    20253.37
]

plt.figure(figsize=(6,4))
plt.bar(models, mae_values)

plt.title("Forecast Accuracy (MAE)")
plt.ylabel("MAE")
plt.show()
```

```python
growth = 10

latest_forecast = forecast_df["Hybrid"].iloc[-1]

adjusted_forecast = latest_forecast * (1 + growth/100)

print("Latest Forecast :", latest_forecast)
print("Growth % :", growth)
print("Adjusted Forecast :", adjusted_forecast)
```
