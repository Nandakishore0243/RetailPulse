```python
import pandas as pd
import numpy as np

from evidently import Report
from evidently.presets import DataDriftPreset
```

```python
rfm = pd.read_csv(
    "../data/processed/rfm_customer_data.csv"
)

rfm.head()
```

```python
reference_data = rfm.copy()

current_data = rfm.copy()

current_data["Monetary"] = (
    current_data["Monetary"] * 1.15
)

current_data["Frequency"] = (
    current_data["Frequency"] + 2
)
```

```python
report = Report(
    metrics=[
        DataDriftPreset()
    ]
)

report.run(
    reference_data=reference_data,
    current_data=current_data
)

print("Drift analysis completed!")
```
