```python
print("""
Day 23 - Kubernetes Deployment Setup

Objective:
- Create Namespace Manifest
- Create Deployment Manifest
- Create Service Manifest
- Prepare RetailPulse for Kubernetes Deployment
""")
```

```python
import os

os.makedirs("../k8s", exist_ok=True)

print("k8s folder created successfully")
```

```python
namespace_yaml = """
apiVersion: v1
kind: Namespace

metadata:
  name: retailpulse
"""

with open("../k8s/namespace.yaml", "w") as f:
    f.write(namespace_yaml)

print("namespace.yaml created")
```

```python
with open("../k8s/namespace.yaml","r") as f:
    print(f.read())
```

```python
deployment_yaml = """
apiVersion: apps/v1
kind: Deployment

metadata:
  name: retailpulse-deployment
  namespace: retailpulse

spec:
  replicas: 1

  selector:
    matchLabels:
      app: retailpulse

  template:
    metadata:
      labels:
        app: retailpulse

    spec:
      containers:
      - name: retailpulse
        image: retailpulse:latest

        ports:
        - containerPort: 8501
"""

with open("../k8s/deployment.yaml", "w") as f:
    f.write(deployment_yaml)

print("deployment.yaml created")
```

```python
with open("../k8s/deployment.yaml","r") as f:
    print(f.read())
```

```python
service_yaml = """
apiVersion: v1
kind: Service

metadata:
  name: retailpulse-service
  namespace: retailpulse

spec:
  selector:
    app: retailpulse

  ports:
  - protocol: TCP
    port: 80
    targetPort: 8501

  type: LoadBalancer
"""

with open("../k8s/service.yaml", "w") as f:
    f.write(service_yaml)

print("service.yaml created")
```

```python
with open("../k8s/service.yaml","r") as f:
    print(f.read())
```

```python
print("""
Day 23 Completed Successfully

Artifacts Created:
- namespace.yaml
- deployment.yaml
- service.yaml

Deployment Ready:
- Kubernetes Namespace
- Kubernetes Deployment
- Kubernetes Service

Status: COMPLETE
""")
```
