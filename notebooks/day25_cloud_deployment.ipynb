```python
import os

os.makedirs("../deployment", exist_ok=True)

print("Deployment folder created successfully")
```

```python
aws_guide = """
# AWS Deployment Guide

1. Create EC2 Instance
2. Install Docker
3. Pull RetailPulse Image
4. Run Container
5. Open Port 8501
6. Access Application

Example:

docker build -t retailpulse .
docker run -p 8501:8501 retailpulse
"""

with open("../deployment/aws_deployment.md","w") as f:
    f.write(aws_guide)

print("AWS deployment guide created")
```

```python
gcp_guide = """
# GCP Deployment Guide

1. Create Compute Engine VM
2. Install Docker
3. Build RetailPulse Container
4. Run Streamlit Application
5. Configure Firewall Rules

Example:

docker build -t retailpulse .
docker run -p 8501:8501 retailpulse
"""

with open("../deployment/gcp_deployment.md","w") as f:
    f.write(gcp_guide)

print("GCP deployment guide created")
```
